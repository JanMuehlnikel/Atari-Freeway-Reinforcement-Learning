```python
import gymnasium as gym
import numpy as np
import tensorflow as tf
import time
from collections import deque
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
```

```python
TOTAL_EPISODES = 1000
RENDER_INTERVAL = 1
EVIROMENT_VARIANT = "ALE/Freeway-v5"
```

```python
for episode in range(TOTAL_EPISODES):
    print(f"EPISODE: {episode}")
    # create env (human to render game and see actions)
    if episode % RENDER_INTERVAL == 0:
        env = gym.make(EVIROMENT_VARIANT, obs_type="ram", render_mode="human", difficulty=0, mode=0) 
    else:
        env = gym.make(EVIROMENT_VARIANT, obs_type="ram", difficulty=0, mode=0) 
        
    env.reset()

    # reset episode variables
    step_count = 0
    accumulated_reward = 0
    while True:
        # update step_count
        step_count += 1

        # abseline: go up every step
        next_state, game_reward, game_done, game_loss_of_live, game_info = env.step(1)

        y_pos = env.ale.getRAM()[14] // 3
        print(y_pos)
        crashed = 1 if env.ale.getRAM()[16] != 255 else 0 # RAM(16) =:= Collision Lane

        if game_reward == 1:
            print("Crossed Line!")
            accumulated_reward += 1

        if game_done:
            print(accumulated_reward)
            env.close()
            break
```
